```python
# importo librerie

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
traffico_aereo = pd.read_excel("traffico-aereo-2012.xlsx")
traffico_aereo.head()
```

```python
traffico_aereo.info()
```

```python
df1 = traffico_aereo.groupby("Aeroporti").agg(
    Pax_commerciale_internazionale = ("Passeggeri  Traffico aereo commerciale di linea internazionale", sum),
    Pax_commerciale_nazionale = ("Passeggeri Traffico aereo di linea nazionale", sum),
    Pax_charter_internazionale = ("Passeggeri Traffico aereo charter internazionale", sum),
    Pax_charter_nazionale = ("Passeggeri Traffico aereo charter nazionale", sum))
df1
```

```python
df1.replace("-",np.nan, inplace = True)
```

```python
df1.isnull().sum()
```

```python
df1 = df1.dropna(how = "all")
df1
```

```python
df1.isnull().sum()
```

```python
df1 = df1.dropna(how ="any")
df1.head(3)
```

```python
df1.isnull().sum()
```

```python
df1["Aeroporto"] = df1.index
```

```python
df1.plot.bar(x = "Aeroporto")
plt.ylabel("Percentuale")
plt.title("CONFRONTO PAX NAZIONALE/INTERNAZIONALE")
```
