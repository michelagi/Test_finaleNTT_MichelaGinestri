```python
# abbiamo parlato anche di valori nulli, questi valori nulli in qualche modo vanno gestiti, essendo tale valori
# una mancanza di informazioni possiamo direttamente eliminarli
nba.tail()
nba.dropna() #rimuove tutta la riga dal dataframe anche se ci sono valori un solo valore nullo, questo è dipeso
# dal parametro how, se tale parametro viene settato a all vengono elimitate le righe se e solo se tutte le occorenze
# sono null.
nba.dropna(how = "all") # se vogliamo rendere effettiva la modifica allo stesso tempo facciamo ovviamente inplace = True
```

# Esercizio ordinamento

```python
# STIAMO VEDENDO GLI ETL PER CUI ORA ANDIAMO A VEDERE LA PARTE DI FILTRAGGIO DEI DATI e di TRASFORMAZIONE
df = pd.read_csv("employees.csv")
df.head(3) # come potete notare in questo dataset abbiamo moltissimi valori diversi se vogliamo vederli meglio
df.info()
# guardando bene la start date e lastlogin potete notare che sono due tipi Object il che porta pandas a vederli stringhe
# questo non va bene perchè se dobbiamo fare operazioni di trasformazioni su differenze possiamo andare in netta difficoltà
# per cui possiamo pensare prima di trasformarli ad esempio
df["Start Date"] = pd.to_datetime(df["Start Date"])
df.head(3)
df["Last Login Time"] = pd.to_datetime(df["Last Login Time"])
```

```python
# iniziamo a filtrare, pensiamo di controllare le persone di sesso femminile
df["Gender"] == "Female"
# se volete visualizzare il tutte le righe che contengono donna
df[df["Gender"]=="Female"]

#oppure
df["Salary"] > 100000
df[df["Salary"]> 100000]
```

```python
# è possibile filtrare anche più di una condizione, inserendo OR e AND
# facciamo finta di filtrare tutti i record dove i lavoratori sono uomini e lavorano nel campo del Marketing
mask1 = df["Gender"] == "Male" 
mask2 = df["Team"] == "Marketing"
df[mask1 & mask2]
# e l'OR?
df [mask1 | mask2]
# potete ovviamente combinare i due operatori logici
```

```python
# è possibile controllare anche se più di un valore è all'interno di una determinata colonna
df["Team"].isin(["Finance","Sales"]) # lo inserite in una maschera
mask3 = df["Team"].isin(["Finance","Sales"])
df[mask3]
```

```python
#metodo between, ci permette di controllare se un determinato valore e in mezzo a n valori decisi da noi
df["Salary"].between(60000,70000)
df["Last Login Time"].between("08:30AM","12:00PM")
```

```python
# altri due metodi importanti possono essere unique e nonunique
df["Team"]
#se voglio sapere rapidamente senza ripetizioni quali sono i team uso unique
df["Team"].unique()
# per contare quanti sono determinati valori aggiungiamo la n avanti a unique
df["Team"].nunique()
```

```python
#abbiamo visto come estrarre informazioni da una colonna, ma come le estraiamo da una riga?
df.loc[5]
# e se vogliamo estrarle da 2 o più righe?
df.loc[[3,4]]
# e se voglio estrarle dalla riga 5 alla riga 9?
df.loc[5:9]

#ora ipotizziamo di volere solo alcune colonne delle righe estratte! Bhe semplicissimo
df.loc[[3,4],["Gender","Team"]]
```
