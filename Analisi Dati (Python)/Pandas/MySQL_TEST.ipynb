```python
# instauro un connettore per la connessione a MySQL
conn = mysql.connector.connect(user='root',password="root",host='localhost',buffered=True,database='mypython')
# Dichiaro un cursore che utilizzerò per eseguire le query e per controllare il risultato 
cursor = conn.cursor()
# 
databases = ("show databases")
cursor.execute(databases)
for (databases) in cursor:
    print (databases[0])
```

```python
query = ("show tables")
cursor.execute(query)
for (query) in cursor:
    print (query[0])
```

```python
create_movie="""CREATE TABLE movies(
   id INT AUTO_INCREMENT PRIMARY KEY,
   title VARCHAR(100),
   release_year YEAR(4),
   genre VARCHAR(100),
   collection_in_mil INT)"""
cursor.execute(create_movie)
```

```python
describe_table = "DESCRIBE movies"
cursor.execute(describe_table)
result = cursor.fetchall()
for row in result:
    print(row)
```
