```python
df = pd.read_excel("Pandas/coalpublic2013.xls")
```

```python
#Calcolo la somma della colonna "Production (short tons)"
sums = df["Production (short tons)"].sum()
print("Sum of values of Production-column:")
print(sums)
```

```python
#Calcolo la media della colonna "Production (short tons)"
media = df["Production (short tons)"].mean()
print("Mean of column Production:")
print(media)
```

```python
#Calcolo il valore massimo e minimo della colonna "Production (short tons)"
max = df["Production (short tons)"].max()

min = df["Production (short tons)"].min()

print("Min of column Production:")
print(min)
print("Max of column Production:")
print(max)
```
