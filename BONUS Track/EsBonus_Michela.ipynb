```python
#Programma pandas per creare un grafico lineare delle azioni di alphabet inc. tra due date
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('alphabet_stock_data__1_.csv')
df.head()
```

```python
first_date = pd.to_datetime('2020-04-01')
last_date = pd.to_datetime('2020-09-29')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= first_date) & (df['Date']<= last_date)

df_a = df.loc[new_df]
df_b = df_a.set_index('Date')
plt.figure(figsize=(8,8))
plt.xlabel("Data",fontsize=16, color='yellow')
plt.ylabel("Azioni", fontsize=16, color='yellow')
 
df_b['Close'].plot(color='blue');
plt.show()
```

```python
#Programma Pandas per insieme di istogrammi compilati di 
#apertura, chiusura, alti e bassi delle azioni tra due date

df[["Open","Close","High","Low"]].plot.hist(stacked = True, bins=150)
plt.ylabel("Frequenze",fontsize=16, color='yellow')
plt.xlabel("Dati",fontsize=16, color='yellow')
```
